### ***Lazy Evaluation Action***

```python
data = [
    ("Alice",25,"New York"),
    ("Bob",30,"Los Angeles"),
    ("Charlie",28,"Chicago")
]

schema = StructType([
    StructField("name", StringType(), True),
    StructField("age", IntegerType(), True),
    StructField("city", StringType(), True)
])

df = spark.createDataFrame(data, schema=schema)
df.show()
```

```python
df = df.filter(col("city")=='New York')
```
