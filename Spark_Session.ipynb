### ***Lazy Evaluation Action***

```python
data = [
    ("Alice",25,"New York"),
    ("Bob",30,"Los Angeles"),
    ("Charlie",28,"Chicago")
]

schema = StructType([
    StructField("name", StringType(), True),
    StructField("age", IntegerType(), True),
    StructField("city", StringType(), True)
])

df = spark.createDataFrame(data, schema=schema)
df.show()
```

```python
df = df.filter(col("city")=='New York')
```

```python
# Action
display(df)
```

###  ***RDD*** = **Resilient Distributed Dataset**

# RDD Explaination
RDD stands for Resilient Distributed Dataset, the core data structure in Apache Spark, representing an immutable, partitioned collection of data spread across a cluster for parallel processing, enabling fault tolerance and fast in-memory computation by tracking lineage for recovery, not data replication. It's "Resilient" (recovers from failures), "Distributed" (across nodes), and a "Dataset" (collection of elements).
